```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import csv
import sys
import time
```

```python
#start time
print('Start Time->',time.ctime())

page_ip = 'http://172.28.64.70:8080'
# home_url = 'http://172.28.64.70:8080/jspui/browse?type=datesubmitted'  # 20 entries per table
home_url = 'http://172.28.64.70:8080/jspui/browse?type=datesubmitted&sort_by=2&order=DESC&rpp=100&etal=0&submit_browse=Update' # 100 entries per table


# next_url_1 = 

Data = pd.DataFrame(columns=['Submission Date','Title','Author', 'Supervisor','Degree','Department','Abstract','URL'])

source = requests.get(home_url).text

soup = BeautifulSoup(source, 'lxml')

total_entries = soup.find('div', class_='panel-heading text-center').text.split(' ')
# print(total_entries)
total_entries = 17985

next_page_url = home_url

for i in range(180):
    source = requests.get(next_page_url).text
    soup = BeautifulSoup(source, 'lxml')    

    table = soup.find('table', class_='table')
    rows = table.find_all('tr')
    rows = rows[1:]
    # rows
    
    for row in rows:
        pd_dict = {}
        elements = row.find_all('td')
        pd_dict['Submission Date'] = elements[0].text
        pd_dict['Title'] = elements[1].text
        pd_dict['Author'] = elements[2].text
        pd_dict['Supervisor'] = elements[3].text
        pd_dict['Degree'] = elements[4].text
        pd_dict['Department'] = elements[5].text
        pd_dict['Abstract'] = 'None'
        
        abstract_url = page_ip + elements[1].a['href']
        
        pd_dict['URL'] = abstract_url
                
        abstract_source = requests.get(abstract_url).text
        abstract_html = BeautifulSoup(abstract_source,'lxml')
        abstract_table = abstract_html.find('table',class_='table itemDisplayTable').find_all('tr')
        for temp in abstract_table:
            if (temp.find('td',class_='metadataFieldLabel').text).split(':')[0] == "Abstract":
                abstract_text = temp.find('td',class_='metadataFieldValue').text
                pd_dict['Abstract'] = abstract_text
    
        Data = Data.append(pd_dict, ignore_index=True)
    
    next_page_url = soup.find('div', class_='panel-heading text-center').find('a', class_='pull-right')['href']
    next_page_url = page_ip+next_page_url
    print(i,next_page_url)

#end_time
print('End Time ->',time.ctime())
```
