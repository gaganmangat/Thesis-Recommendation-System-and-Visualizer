```python
import numpy as np
import matplotlib.pyplot as plt
import csv
import re
from nltk.corpus import stopwords  
from nltk.tokenize import word_tokenize  
from sklearn.feature_extraction.text import TfidfVectorizer
import scipy.sparse
from numpy import dot
from numpy.linalg import norm
```

```python
data=[]
# filename="/home/rohitk07/Downloads/data_project.csv"
filename="../Pre-Processing/data_cleaned.csv"
with open(filename, 'r', encoding='utf-8') as csvfile: 
    # creating a csv reader object 
    csvreader = csv.reader(csvfile) 
      
    # extracting field names through first row 
    fields = next(csvreader) 
  
    # extracting each data row one by one 
    for row in csvreader: 
        data.append(row) 
```

```python
z=[]
id={}
for x in data:
    x[6]=x[6].replace('\n'," ")
    z.append(x)
    id[x[1]]=int(x[0])
data=z
```

```python

def remove_stopwords(text):
    
      stop_words = set(stopwords.words('english')) 
      #text = re.sub(r'[^\w\s]', '', text)         
      words = text.split()
      ans = ""
      for w in words:
          if w not in stop_words:
                ans=ans+ " "+w
                
      return ans          



def remove_punctuation(text):
    punc = '''!()-[]{};:'"`\, <>./?@#$%^&*_~'''
    text=list(text)   
    for i in range(len(text)):      
           if text[i] in punc:
                text[i]=" " 
    return "".join(text)            
                
                
```

```python
abstracts = [x[6] for x in data]

print(remove_punctuation("Zero-shot learning"))
```

```python
# cleaning the data

from nltk.stem import PorterStemmer 
from nltk.tokenize import word_tokenize 
z=[]
for x in abstracts:
    x = remove_punctuation(x)
    x = remove_stopwords(x)
    z.append(x)
abstracts_cleaned=z    
            
```

```python
#vectorizer = TfidfVectorizer(min_df=3,max_features=None, strip_accents='unicode',analyzer='word',token_pattern=r'\w{1,}',ngram_range=(1,3) ,stop_words='english' )
vectorizer  = TfidfVectorizer()
```

```python
vectorizer.fit(abstracts_cleaned)
```

```python
print(vectorizer.idf_)
```

```python
vector=vectorizer.transform(abstracts_cleaned)
```

```python
scipy.sparse.save_npz("vector.npz",vector)
```
